```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rc('font', family='gulim')
mpl.rc('axes', unicode_minus = False)
```

```python
plt.title("제목")
plt.plot([10, 20, 30, 40], [1, 7, 9, 16], 'gx:')
plt.xlabel("x축 라벨")
plt.ylabel("y축 라벨")
plt.show()
```

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러 개의 선 그리기")
plt.plot(
    t, t, 'r--', 
    t, 0.5 * t**2, 'bs:',
    t, 0.2 * t**3, 'g^-'
)
plt.show()
```
