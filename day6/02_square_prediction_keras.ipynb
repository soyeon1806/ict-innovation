```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
x_train = np.array([-12, -10, -8, -6, -4, -2, 2, 4, 6, 8, 10, 12])
y_train = np.array([144, 100, 64, 36, 16, 4, 4, 16, 36, 64, 100, 144])
```

```python
model = Sequential()
model.add(Dense(64, input_dim = 1, activation = 'relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(loss = 'mse', optimizer = Adam(learning_rate = 0.01))
model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs = 500)
print(hist.history.keys())
```

```python
x_test = np.array([-11, -5, -3, 3, 5, 7, 10])
predictions = model.predict(x_test)

for i, value in enumerate(x_test):
    print(f"Input: {value}, Predicted Square: {predictions[i][0]}")

plt.plot(x_train, y_train, 'bo', label='Actual')
plt.plot(x_test, predictions, 'r+', label='Predicted')
plt.legend()
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Actual vs Predicted Values')
plt.show()
```
